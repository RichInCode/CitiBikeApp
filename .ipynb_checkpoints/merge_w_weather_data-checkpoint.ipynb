```python
class line:
    def __init__(self, m=None, b=None):
        self.m = m
        self.b = b
    def calculateMandB(self, x1, y1, x2, y2):
        self.m = (y2 - y1)/(x2 - x1)
        self.b = y1 - self.m*x1
    def getY(self, x):
        return self.m*x + self.b
```
